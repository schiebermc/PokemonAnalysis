```python
# Here is a separate notebook for formatting the data how I want. I pulled the original dataset from Kaggle.
import pandas as pd
pd.options.display.max_columns = None
```

```python
pokemon = pd.read_csv('pokedex_(Update.04.20).csv')
```

```python
pokemon.head()
```

```python
# list of columns
pokemon.columns
```

```python
# here I am dropping some columns that I do not anticipate to being too interesting. 
pokemon.drop(columns=['Unnamed: 0', 
                      'percentage_male', 
                      'german_name', 
                      'japanese_name', 
                      'base_friendship', 
                      'egg_type_number', 
                      'egg_type_1',
                      'egg_type_2',
                      'percentage_male',
                      'egg_cycles',
                     ], inplace=True)
```

```python
# One big reason I am pulling in this larger dataset is to get abilities info such as Huge Power
pokemon[pokemon['name'] == "Azumarill"]
```

```python
# here I want to transform the name to a captilized type for consistency with my other dataset
mapped_names = {}
for column in pokemon.columns:
    if "against_" in column:
        name = column.split('_')
        new_col = name[0] + '_' + name[1].capitalize()
        mapped_names[column] = new_col
mapped_names
```

```python
pokemon.rename(axis=1, mapper=mapped_names, inplace=True)
```

```python
pokemon.head()
```

```python
# write the transformed data to a new file. We will use this in the "Analysis" notebook
pokemon.to_csv("Pokemon_extensive.csv")
```
